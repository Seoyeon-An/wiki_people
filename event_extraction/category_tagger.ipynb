```python
import pandas as pd

def extract_column_as_list(file_path, column_index):
    df = pd.read_csv(file_path)
    if column_index < len(df.columns):
        column_data = df.iloc[:, column_index].tolist()
        return column_data
    else:
        print(f"Column index {column_index} is out of range.")

result_list = extract_column_as_list(r'C:\Users\inny9\wiki_people\event_extraction\Winston_Churchill.csv',2)
```

```python
import yake
from time import time

language = "en"
max_ngram_size = 3
deduplication_threshold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 1

all_elements = result_list

start_time = time()

for text in all_elements:
    custom_kw_extractor = yake.KeywordExtractor(lan=language,
                                                n=max_ngram_size,
                                                dedupLim=deduplication_threshold,
                                                dedupFunc=deduplication_algo,
                                                windowsSize=windowSize,
                                                top=numOfKeywords,
                                                features=None)
    keywords = custom_kw_extractor.extract_keywords(text)

    print(f"Keywords for text: '{text}'")
    for kw, score in keywords:  # lower the score, more relevant the keyword
        print(f"{kw} (Score: {score:.4f})")
    print("\n")

end_time = time()
running_time = end_time - start_time

print("model running time: %s " % f"{running_time:12.9f}", end="")
```
