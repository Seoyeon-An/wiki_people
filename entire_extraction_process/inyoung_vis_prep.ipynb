```python

import pandas as pd
import glob
import os

# Path to the directory containing CSV files
directory_path = r'C:\Users\inny9\Documents\GitHub\wiki_people\data\final_df\*.csv'  # Replace 'your_directory_path' with your directory path
output_file = 'merged_file.csv'  # Name of the output file

# Get a list of file paths for all CSV files in the directory
file_list = glob.glob(directory_path)

for file in file_list:
    file_name = os.path.basename(file)  # Extract the file name without the path
    df = pd.read_csv(file)  # Read the CSV into a DataFrame
    df['Name'] = str(file_name)[:-4]
    df.to_csv(file, index=False)  # Save the modified DataFrame back to the CSV file
```

```python
# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Loop through each file and concatenate data into the merged_data DataFrame
for file in file_list:
    data = pd.read_csv(file)  # Read each CSV file
    merged_data = pd.concat([merged_data, data], ignore_index=True)  # Concatenate data

# Write the merged data to a new CSV file
merged_data.to_csv(output_file, index=False)

df = pd.read_csv(output_file)
df

```

```python
new_column = df['Name']  # Extracting the new column

# Dropping the 'New_Column' from its current position
df = df.drop(columns=['Name'])

# Inserting the 'New_Column' at index 0
df.insert(loc=0, column='Name', value=new_column)

df
```

```python
# Create a new column based on matching keys from 'Column_0' to the dictionary keys
df['Occupation'] = df['Name'].map(people_occupations)
#df = df.drop('Occupation', axis=1)
df

```

```python
condition = range(2111, 2128)
df.loc[condition, 'Name'] = df.loc[condition, 'Name'].apply(lambda x: x[:-1])
```

```python
#check if the occupations are all mapped correctly without omissions
nan_count = df['Occupation'].isna().sum()

# Display the count of NaN values in 'Column_A'
print(f"Number of NaN values in 'Occupation': {nan_count}")

rows_with_nan = df[df['Occupation'].isnull()]
print("Rows with NaN in 'Occupation':")
rows_with_nan
```
